2 - Functions

Function Type Syntax
In the function signature, variables can statically be declared with a type expectation.

Example:
function createMessage(name: string, a: number, b: number): string {
    return `${name} score ${a + b}`;
}
and with arrow functions...
function createMessage = (name: string, a: number, b: number) string => {
    return `${name} scored ${a + b}`;
}

*utility functions (calculations, data retrival, etc.) within typescript are generally stored in utils.ts*

Inferred Return Types
- Implicit return types are more accurate/narrow generally
- Intention matters
- Explicit return when many output value types are possible to specify expectations
- When only one outcome is expected, inferred is preferred.
- Code used by self: implicit, by others: explicit.

Example:
Taking two strings, a string value is inferred as a return type...

export function combinePrompts(systemPrompt: string, userPrompt: string) {
  return systemPrompt + "\n" + userPrompt;
}

Void
- Represents/communicates intent that the return value of functions that don't return a value
- undefined is returned by default. void ensures that truly nothing is returned.

Function Types
- Includes info about params and return value
Examples:
*(param1: type1, param2: type2, ...) => returnType

*(a: number, b: number) => number;

*const add = (a, b) => a + b;
const subtract = (a, b) => a - b;

Type Alias
- Avoids long custom typing
- As long as the signature matches, the function conforms to the custom type.
Example:
export type SupportResponse = (name: string) => string;

export function greetCustomer(name: string) {
  return `Hello ${name}, welcome to Support.ai! How can I assist you today?`;
}

export function farewellCustomer(name: string) {
  return `Goodbye ${name}, have a great day!`;
}

Importing Types
- Certain TypeScript configs can allow types to be imported directly from modules
    import { User, Post } from "./models";
- Safer and more efficient to use import type
    import type { User, Post } from "./models";
    
    or

    import { type User, type Post } from "./models";

- Communicates that you are only importing types. Drops unncessary import bundle size, avoiding extra JavaScript code.
