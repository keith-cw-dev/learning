```python
Lambda Expressions mini quiz

Q1 – Sorting with a custom key
You have a list of user records:

users = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35},
]

You want to sort by age once in this file and you won’t reuse the “get age” logic anywhere else.

You could do:

def get_age(user):
    return user["age"]

users.sort(key=get_age)

or:

users.sort(key=lambda user: user["age"])

In this situation, which is more appropriate and why? (Pick one and justify in a sentence or two.)
For question one, I would personally go with the one line lambda expression as it both collects the ages per user
and sorts them in the same line and sorts as unique values are found. The alternative seems to take much longer by grabbing each specific unique value, and then iterating through the collected list of values to sort

Q2 – Reusing logic across the codebase
Suppose you frequently need to normalize usernames in many places:

def normalize_username(name):
    # complicated logic, used in many modules
    ...

Would you rather implement that as:

A) a top-level def normalize_username(...)
B) a lambda assigned to a name: normalize_username = lambda name: ...
Which would you choose, and why, in terms of readability, debugging, and maintainability?

Q3 – Mapping vs loop
You have:

numbers = ["10", "101", "111"]

You want a new list of int values from binary strings.

Option 1:

result = [int(n, 2) for n in numbers]

Option 2:

result = list(map(lambda n: int(n, 2), numbers))

Assuming both are equally fast, which would you prefer in idiomatic Python and why?
I would go with list(map(lambda n: int(n, 2), numbers)) as it reads easier for me and would allow references to the old values just in case I need to use them in the future.

Answer these three, and then I’ll:

Correct them,
Point out which scenarios are truly “good lambda” vs “bad lambda,”
Give you 1–2 patterns where lambdas genuinely shine.
________
Good thinking, young coder-bear, but there are a couple of important things you’re mixing together. Let’s untangle them.

You said:

the one line lambda expression … collects the ages per user and sorts them in the same line … The alternative seems to take much longer …

Key points you’re overlooking:

Performance is the same in practice
def get_age(user):
    return user["age"]

users.sort(key=get_age)

and

users.sort(key=lambda user: user["age"])

Both:

Call a small function once per element,
Let sort handle the same list with the same algorithm.
Python doesn’t first “collect ages” into a separate list in either case; it just calls the key function on demand. Whether that key is a def or a lambda doesn’t change the sorting behavior.

So: no real efficiency gain. It’s a micro-level difference at best, usually noise.

“Unique values” don’t change what key does
sort(key=...):

Calls the key function for every element,
Gets one key per element,
Sorts based on those keys.
It doesn’t skip duplicates or do extra passes because of def vs lambda. The algorithm is the same.

The real tradeoff is clarity, not speed
lambda user: user["age"] is great when:
The logic is tiny,
It’s used once,
The meaning is obvious from context.
def get_age(user): ... is better when:
You want to reuse it,
It needs a docstring or comments,
You want clearer stack traces and names in debugging.
So for Q1, your reasoning about performance is off, but your choice (using a lambda there) is still reasonable: it’s a short, one-off key function.

If we rephrase your answer in “Pythonic” terms, it would be:

I’d choose the lambda because it’s a tiny, one-off transformation used only as the key, so defining a separate named function would add noise without improving clarity.

Treat lambda as “a tiny, throwaway function” for now.
Prefer a normal def whenever:
- You’ll reuse the function,
- It’s more than a one-liner,
- You want to document or debug it easily.

Nested Loops - Multi-iteration versus single passes

Building an extension dictionary by type mapping - Single pass, nested
file_extension_tuples = [
    ("document", [".doc", ".docx"]),
    ("image", [".jpg", ".png"]),
]

ext_to_type = {}
for file_type, extensions in file_extension_tuples:
    for ext in extensions:
        ext_to_type[ext] = file_type

Processing cells in a grid - Multi-pass, nested
grid = [
    [1, 2, 3],
    [4, 5, 6],
]

for row in range(len(grid)):
    for col in range(len(grid[row])):
        value = grid[row][col]
        # process value



Zip in Python:
for x, y in zip(xs, ys):
    print(x, y)

Enumerate in Python:
for i, value in enumerate(values):
    print("index:", i, "value:", value)

First-Class and Higher Order Functions

First-Class: treated like any other value
example:
def square(x):
    return x * x

# Assign function to a variable
f = square

print(f(5))
# 25

Higher Order: function that accepts function(s) and returns function(s)
example:
def square(x):
    return x * x

def my_map(func, arg_list):
    result = []
    for i in arg_list:
        result.append(func(i))
    return result

squares = my_map(square, [1, 2, 3, 4, 5])
print(squares)
# [1, 4, 9, 16, 25]

Filter Function exercise review
My Solution:
def remove_invalid_lines(document):
    lines = document.split("\n")
    for line in lines:
        if line.startswith("-"):
            lines.remove(line)
    return ("\n").join(lines)
    
Optimal Solution:
def remove_invalid_lines(document):
    return "\n".join(
        filter(lambda line: not line.startswith("-"), document.split("\n"))
    )
```
