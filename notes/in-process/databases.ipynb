```python
9 - Normalization

Types of Relationships
One - to - One; Add a new column
One - to - Many; Two tables needed w/ primary keys and referenced appropriate table
Many - to - Many; Add a third joining table

CONSTRAINT constraint_name
FOREIGN KEY (secondary_table_variable)
REFERENCES (main_table_variable)

*MAPPING TABLES*

Joining Tables
UNIQUE keys can be declared with FOREIGN KEY and REFERENCES after
example:
CREATE TABLE product_suppliers (
  product_id INTEGER,
  supplier_id INTEGER,
  UNIQUE(product_id, supplier_id),
  FOREIGN KEY (product_id) REFERENCES products (id),
  FOREIGN KEY (supplier_id) REFERENCES suppliers (id)
);

or

CREATE TABLE users_countries (
  country_id,
  user_id,
  UNIQUE(country_id, user_id),
  FOREIGN KEY (country_id) REFERENCES countries (id),
  FOREIGN KEY (user_id) REFERENCES users (id)
);

Normal Forms

"Simplest form, no copies"

1NF
- Every row must have unique primary keys
- No nested tables

2NF
- All columns not a part of primary key (smallest number
of columns that can be used to identify a row), must be dependent on primary key
(can fix with a mapping table)

3NF
- Columns not in primary key must be dependent only on primary key
(can fix also with a mapping table)

Boyce Codd NF
- Column part of primary key may not be dependent on column not a part of primary key

Composite Keys - 2 or more columns that serve as primary keys



```
