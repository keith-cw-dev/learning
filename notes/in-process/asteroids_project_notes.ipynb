Learning Goals:
- To introduce you to multi-file Python projects
- To get you used to referencing documentation (something you'll do often in the real world)
- To show you a real-world use case for object-oriented programming
- To have a ton of fun building a rewarding project!

As a best practice, each Python project on your machine should have its own virtual environment to keep them isolated from each other.
(.venv)

*make sure python, pygame, and uv are up to date, as well as any other dependencies*

uv init your-project-name
- initializes project directory. do from home directory

uv python pin 3.x
- pin a version of a dependency via uv

uv venv
- creates a virtual environment (venv). best
done at the top level of project directory.

source .venv/bin/activate
- activates the venv

uv add denpendency==x.x.x (dependency-version)
- adds a dependency with a specific version

uv run -m dependency
- ensures dependency is installed
