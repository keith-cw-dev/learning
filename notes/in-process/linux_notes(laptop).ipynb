```python
sudo 
- a preface that temporarily elevates permissions for a command

ls
- lists the contents of a dir

cat (concatenate)
- displays the full contents of a file

head -n ***#oflines*** filename.txt 
- reads the file from the beginning by n number of lines.

tail -n ***#oflines*** filename.txt
- reads the file from the end by n number of lines

cd .. 
- backs out of file path by 1 layer

cd /dir/path
- changes directory to location of direct path
- can also change relatively if already at necessary folder location via cd ***foldername***

touch filename.txt
- creates new file if not existing, makes access logs for files

rm filename.txt
- removes a given file

mkdir directoryname
- makes a subdirectory in current directory

mv file.txt new_name.txt
- changes file name

mv some_file.txt some_dir/some_file.text
- moves file from current dir to another nested dir

mv some_file.txt ../some_file.txt
- file from current to parent dir

mv some_file.txt some_directory/
(can also omit filename when moving, unless renaming as well)

cp source_file.txt destination/
- copy a file from one place to another

cp -r dir_one dir_two
- copy the contents of one directory into another, recursively

cd ~
-returns to home directory

grep "string_value" filename.ext
- searches for a given string value in a given file

grep -r "string_value" .
- recursive search through current dir and subdirs

. == alias for "current directory"

* == "wildcard" character/filler 
example:
# Find all filenames that contain the word "chad"
find some_directory -name "*chad*"

find some_directory -name hello.txt
- find by filename

find some_directory -name "*.txt"
- find all files with given pattern

permissions are stored as 10-char strings:
first *: - or d (file or dir)
2-4 *: rwx owner (read, write, and execute); will have dashes where perms are not allowed
5-7 *: rwx group
8-10 *: rwx others

shebang - line at top of script that signals the shell what program to execute
format: #! interpreter [optional-arg]

example: 
#!/usr/bin/python3
This tells the system to use the Python 3
interpreter located at /usr/bin/python3 to run the script.

setting variables in shell:
export NAME="Lane"
- programs and scripts can also use the declared variables
- i.e:
#!/bin/sh
echo "Hi I'm $NAME" # can create new file and use the variable set via export

chmod +x ./introduce.sh # and then run it with the utilized variable

./introduce.sh
# Hi I'm Lane

echo $PATH
- shows all directories stored in PATH

export PATH=""
- resets PATH to be an empty string (useful for making commands more explicit)
- only affects current shell session

export PATH="$PATH:/path/to/new"
- adds a new PATH directory without overwriting others

nano ./~bashrc
- allows config editing for bash source file

source ./~bashrc
- resets source for shell session/effectively sestarts shell session

flags
single character -
multi character --
- some flags can utilize both (-h and -help)

exit codes
0 - success
1 - "catch-all" code

ls ~
echo $?
# 0

ls /does/not/exist
echo $?
# non-zero (depends on your OS)
- access exit code of last program ran









```
