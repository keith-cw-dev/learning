```python
sudo apt install pandoc
- document format converter

git switch -c my_new_branch
- switches to branch unless it does not exist, in which git creates the branch using -c flag

git log --oneline --decorate --graph --parents
- more visual representation of merge flows

This is a common workflow when working with Git on a team of developers:

- Create a branch for a new change
- Make the change
- Merge the branch back into main (or whatever branch your team dubs the "default" branch)
- Remove the branch
- Repeat

git branch -d name_of_branch
- deletes a branch

git switch -c update_dune COMMITHASH
- creates a new branch using a prev branch hash
 for relevant content

git rebase branch_name
- rebase to bring changes from one branch to current

cd to home dir > mkdir dir_name-local >
 cd dir_name-local > git init
- establishes a local directory

git remote add <name> <uri>
- makes a remote repo under an alias '<name>' that starts
 from local file locations <uri [Uniform Resource Identifier]>

find .git/objects
- to retrieve data and contents info from given directory

git fetch 
- downloads copies of all the contents
 of found directory data/info

git fetch origin
- downloads copies of all the contents of the remote directory (if origin
is used as name for remote command)

git log [<remote>/<branch>]
- creates a log of commits for given branch

git merge remote/branch
- merges commits from remote repo's branch

git push origin main
- pushes local changes to a given remote repo

git push origin <localbranch>:<remotebranch>
- push a local branch to a remote with a different name

git push origin :<remotebranch>
- push an empty branch to delete a remote branch

Core Solo Workflow
- Make changes to files
- git add . (or git add <files> if I only want to add specific files)
- git commit -m "a message describing the changes"
- git push origin main

Core Team Workflow
- Update my local main branch with git pull origin main
- Checkout a new branch for the changes I want to make with git switch -c <branchname>
- Make changes to files
- git add .
- git commit -m "a message describing the changes"
- git push origin <branchname> (I push to the new branch name, not main)
- Open a pull request on GitHub to merge my changes into main
- Ask a team member to review my pull request
- Once approved, click the "Merge" button on GitHub to merge my changes into main
- Delete my feature branch, and repeat with a new branch for the next set of changes


creating patterns in .gitignore
*.txt
!/important.txt
* = inclusive of all files with given patterns
! = negates pattern
/ = specifies patterns anchored to directory 
containing .gitignore file
# = comments in .gitignore file

Order matters!
temp/*
!temp/instructions.md

Everything in the temp/ directory would be ignored except 
for instructions.md. If the order were reversed, instructions.md 
would be ignored.

What to ignore:
- Ignore things that can be generated
(e.g. compiled code, minified files, etc.)
- Ignore dependencies (e.g. node_modules, venv, packages, etc.)
- Ignore things that are personal or specific 
to how you like to work (e.g. editor settings)
- Ignore things that are sensitive or dangerous 
(e.g. .env files, passwords, API keys, etc.)

git rm --cached file
- remove a staged or commited file





```
