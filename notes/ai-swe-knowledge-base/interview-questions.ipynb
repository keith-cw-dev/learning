This notebook contains curated interview questions and structured answers.

Goal:
Build fast recall + structured explanation skills for:
- SWE interviews
- AI evaluator interviews (Mercor / Alignerr / Outlier)
- Live technical explanation under pressure

Answer Structure:

1. Direct Answer (1–2 sentences)
2. Intuition / Mental Model
3. Technical Breakdown
4. Example
5. Edge Case or Tradeoff

Common Interview Mistakes:

- Over-explaining without structure
- Jumping to implementation too fast
- Ignoring edge cases
- Confusing similar concepts (e.g. useMemo vs useCallback)
- Using jargon without explanation

Prompt:
Explain [TOPIC] in 60–90 seconds.

Evaluation Criteria:
- clarity
- structure
- correctness
- simplicity
- real-world analogy
