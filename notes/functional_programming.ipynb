# Functional Programming

1 - What is Functional Programming?

Style of programming that describes what is wanted, not how to do it. Computation is treated as the evaluation of mathematical functions.
Declarative and does not require continuous re-updating. Makes code easier to test, debug, and think about. Not necessarily about "code golfing (aiming for as little lines of code as possible).

"Apply this transformation to this data"

Key Characteristics:


Functional Example: 
return enable_4wd(clean_windows(add_gas(create_car())))

# Imperative Programming - opposite of functional programming
Style of programming where the computer is instructed what to do with the data step by step. Explicit commands are given to change the state of the program over time (recipe with exact instructions).

Imperative exmaple:
car  = create_car()
car.add_gas(10)
car.clean_windows()
car.four_wd = true
- at every step, the car is updated. Time must be considered.

Key characteristics:
- Statements that change state (variables get updated)
- Control flow: loops (for, while), conditionals (if)
- Focus on how the result is achieved
- Order of execution matters

Common Imperative Languages:
- Objective-C
- Java
- Python (has multi-paradigms)
- JS
- C++

# 5 - Function Transformations

Specific type of higher order functions that take other functions as input and return a new function as output.
A function that takes a mathematical operation (like add or multi) and returns a new function that applies that operation twice.
Example:
def multiply(x, y):
    return x * y

def add(x, y):
    return x + y

# self_math is a higher order function
# input: a func that takes 2 args and returns a value
# output: a new func that takes 1 arg and returns a value

def self_math(math_func):
    def inner_func(x)
        return math_func(x, y)
    return inner_func

square_func = self_math(multiply)
double_func = self_math(add)

print(square_func(5))
# Prints 25

print(double_func(5))
# Prints 10

* Best to use function transformations when you want to create specialized versions of existing functions without modifying the original functions and to obtain more understandable and extendable code.

* Use advanced techniques like closures and decorators to create function transformations that can maintain state or add functionality to existing functions to make code simpler than it otherwise would be.

* Dynamic variations of functions can be created using function transformations to handle different input types or behaviors based on runtime conditions, improving code flexibility and adaptability.


# 6 - Closures

A closure is a function that retains access to its lexical scope, even when the function is executed outside that scope.
Closures are created when a nested function references variables from its enclosing scope.
Closures are useful for data encapsulation, maintaining state, and creating functiuon factories.
Example of a closure in Python:
def out_function(x):
    def inner_function(y):
        return x + y
    return inner_function
closure_func = outer_function(10)
print(closure_func(5)) # prints 15
